## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
# ================================
# CELL 1: Import libraries & set configs
# ================================

# basic python libs
import os
import numpy as np
import pandas as pd

# tensorflow / keras for deep learning
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16, ResNet50

# sklearn for metrics and utilities
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight

# plotting
import matplotlib.pyplot as plt
import seaborn as sns

# some global constants
IMG_SIZE = 224          # we'll resize all images to 224x224
BATCH_SIZE = 32         # mini-batch size for training
NUM_CLASSES = 5         # DR levels: 0,1,2,3,4
AUTOTUNE = tf.data.AUTOTUNE  # lets TF optimize data pipeline

print("Libraries imported successfully ")
````

````python
# ================================
# CELL 2: Load train.csv and create image paths
# ================================

# paths where you uploaded the dataset in Colab
CSV_PATH = "/content/train_1.csv"
IMG_DIR = "/content/train_images/train_images"   # folder containing PNG images

# read the csv file
df = pd.read_csv(CSV_PATH)

# create full file path for each image
# example: id_code = "000c1434d8d7" -> "/content/train_images/000c1434d8d7.png"
df["filepath"] = df["id_code"].apply(lambda x: os.path.join(IMG_DIR, x + ".png"))

# make sure label is integer
df["label"] = df["diagnosis"].astype(int)

# show first few rows just to verify
print(df.head())

# check class distribution (important: dataset is imbalanced)
print("\nClass distribution:")
print(df["label"].value_counts())
````

````python
# ================================
# CELL 3: Split into train and validation sets
# ================================

# we use stratify so that class proportions are preserved in both splits
train_df, val_df = train_test_split(
    df,
    test_size=0.2,            # 80% train, 20% validation
    stratify=df["label"],     # keep label distribution similar
    random_state=42
)

print("Number of training samples:", len(train_df))
print("Number of validation samples:", len(val_df))
````

````python
# ================================
# CELL 4: Define TF data pipeline
# ================================

def decode_img(path, label):
    """
    Given:
        path  -> path to the image file
        label -> integer class label
    This function:
        1. reads the file from disk
        2. decodes PNG into a 3-channel image
        3. resizes it to IMG_SIZE x IMG_SIZE
        4. normalizes pixel values to [0,1]
    Returns:
        image tensor, label
    """
    # read the binary file
    img = tf.io.read_file(path)
    # decode as PNG
    img = tf.image.decode_png(img, channels=3)
    # resize
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    # normalize to [0,1]
    img = img / 255.0
    return img, label


def make_dataset(df, shuffle=True, augment_layer=None):
    """
    Create a tf.data.Dataset from a dataframe with 'filepath' and 'label' columns.

    If augment_layer is given, we apply that on the fly to images (for training).
    """
    paths = df["filepath"].values
    labels = df["label"].values

    # create a dataset of (path, label)
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    # map paths to actual images
    ds = ds.map(decode_img, num_parallel_calls=AUTOTUNE)

    # apply augmentation only if provided (for training set)
    if augment_layer is not None:
        ds = ds.map(lambda x, y: (augment_layer(x, training=True), y),
                    num_parallel_calls=AUTOTUNE)

    # shuffle only for training
    if shuffle:
        ds = ds.shuffle(buffer_size=len(df))

    # batch and prefetch for better performance
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds


# create train and val datasets (no augmentation for now)
train_ds = make_dataset(train_df, shuffle=True)
val_ds   = make_dataset(val_df, shuffle=False)

print("TF datasets created ")
````

````python
# ================================
# CELL 5: Evaluation function
# ================================

def evaluate_model(model, val_df, name="Model"):
    """
    Runs the model on the validation set and prints:
    - Accuracy
    - Macro Precision
    - Macro Recall
    - Macro F1 Score

    Returns these metrics so we can store them in tables.
    """
    # create a small dataset only for evaluation (no shuffle, no aug)
    ds = tf.data.Dataset.from_tensor_slices((val_df["filepath"], val_df["label"]))
    ds = ds.map(decode_img).batch(BATCH_SIZE)

    # run the model on the validation data
    preds = model.predict(ds)
    # get predicted class index (0..4) by taking argmax of softmax output
    y_pred = preds.argmax(axis=1)
    # true labels
    y_true = val_df["label"].values

    # compute all metrics using sklearn
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average="macro", zero_division=0)
    rec  = recall_score(y_true, y_pred, average="macro", zero_division=0)
    f1   = f1_score(y_true, y_pred, average="macro", zero_division=0)

    # nicely print them
    print(f"\n {name} Metrics:")
    print(f"Accuracy :  {acc:.4f}")
    print(f"Precision:  {prec:.4f}")
    print(f"Recall   :  {rec:.4f}")
    print(f"F1 Score :  {f1:.4f}")

    return acc, f1, rec, prec
````

````python
# ================================
# CELL 6: Function to build classifier on top of a backbone
# ================================

def build_classifier(base_model):
    """
    Takes a CNN backbone (e.g., VGG16/ResNet50) and adds:
        - GlobalAveragePooling2D
        - Dense(256, relu)
        - Dropout(0.5)
        - Dense(NUM_CLASSES, softmax)
    Also freezes the backbone for baseline models.
    """
    # freeze all layers in the backbone (no training)
    base_model.trainable = False

    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base_model(inputs, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    return model
````

````python
# ================================
# CELL 7: Baseline Model 1 - VGG16 (frozen backbone)
# ================================

# load VGG16 backbone without the original classification head
vgg_base = VGG16(weights="imagenet", include_top=False,
                 input_shape=(IMG_SIZE, IMG_SIZE, 3))

# attach our custom head
vgg_model = build_classifier(vgg_base)

# compile the model
vgg_model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# train for a few epochs
history_vgg = vgg_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=8
)

# evaluate on validation set
vgg_results = evaluate_model(vgg_model, val_df, name="VGG16 Baseline")
````

````python
# ================================
# CELL 8: Baseline Model 2 - ResNet50 (frozen backbone)
# ================================

res_base = ResNet50(weights="imagenet", include_top=False,
                    input_shape=(IMG_SIZE, IMG_SIZE, 3))

res_model = build_classifier(res_base)

res_model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_res = res_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=8
)

res_results = evaluate_model(res_model, val_df, name="ResNet50 Baseline")
````

````python
# ================================
# CELL 9: Baseline Model 3 - AlexNet-like CNN (from scratch)
# ================================

def alexnet_backbone():
    """
    Simple AlexNet-like backbone (no pretrained weights).
    This is just to follow the assignment requirement.
    """
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = layers.Conv2D(96, 11, strides=4, activation="relu", padding="same")(inputs)
    x = layers.MaxPooling2D(3, strides=2)(x)
    x = layers.Conv2D(256, 5, activation="relu", padding="same")(x)
    x = layers.MaxPooling2D(3, strides=2)(x)
    x = layers.Conv2D(384, 3, activation="relu", padding="same")(x)
    x = layers.Conv2D(384, 3, activation="relu", padding="same")(x)
    x = layers.Conv2D(256, 3, activation="relu", padding="same")(x)
    x = layers.MaxPooling2D(3, strides=2)(x)
    return models.Model(inputs, x)

alex_base = alexnet_backbone()
alex_model = build_classifier(alex_base)

alex_model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_alex = alex_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=8
)

alex_results = evaluate_model(alex_model, val_df, name="AlexNet Baseline")
````

````python
# ================================
# CELL 10: Plot training/validation accuracy (for bias-variance discussion)
# ================================

def plot_history(history, title):
    plt.figure()
    plt.plot(history.history["accuracy"], label="train_acc")
    plt.plot(history.history["val_accuracy"], label="val_acc")
    plt.title(title + " Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()

plot_history(history_vgg, "VGG16 Baseline")
plot_history(history_res, "ResNet50 Baseline")
plot_history(history_alex, "AlexNet Baseline")
````

````python
# ================================
# CELL 11: Data augmentation & class weights
# ================================

# data augmentation layer used only during training
data_aug = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.08),
    layers.RandomZoom(0.1),
    layers.RandomContrast(0.1),
], name="data_augmentation")

# compute class weights to handle class imbalance
class_weights_array = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(train_df["label"]),
    y=train_df["label"].values
)

class_weights = {i: w for i, w in enumerate(class_weights_array)}
print("Class weights:", class_weights)
````

````python
# ================================
# CELL 12: Modified model - Stage 1 (train head with frozen ResNet50 + aug + class weights)
# ================================

# load ResNet50 base
ft_base = ResNet50(weights="imagenet", include_top=False,
                   input_shape=(IMG_SIZE, IMG_SIZE, 3))

ft_base.trainable = False  # freeze backbone initially

# build model with augmentation + new head
inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = data_aug(inputs, training=True)   # augmentation happens here
x = ft_base(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(512, activation="relu")(x)
x = layers.Dropout(0.5)(x)
outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

ft_model = models.Model(inputs, outputs, name="ResNet50_Modified")

ft_model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# training dataset (NO AUG in dataset)
train_ds_aug = make_dataset(train_df, shuffle=True)

history_stage1 = ft_model.fit(
    train_ds_aug,
    validation_data=val_ds,
    epochs=5,
    class_weight=class_weights
)

print("Stage 1 training done ")

````

````python
# ================================
# CELL 13: Modified model - Stage 2 (fine-tune top ResNet50 layers)
# ================================

# decide how many layers to unfreeze (fine-tune)
fine_tune_at = 100  # typical choice: unfreeze last ~100 layers

for layer in ft_base.layers[:fine_tune_at]:
    layer.trainable = False
for layer in ft_base.layers[fine_tune_at:]:
    layer.trainable = True

# recompile with LOWER learning rate (important for fine-tuning)
ft_model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_stage2 = ft_model.fit(
    train_ds_aug,
    validation_data=val_ds,
    epochs=10,
    class_weight=class_weights
)

# final evaluation
ft_results = evaluate_model(ft_model, val_df, name="Modified Fine-Tuned ResNet50")
````

````python
# ================================
# CELL 14: Confusion matrix for modified model (nice to show in report)
# ================================

# create validation dataset
val_ds_eval = tf.data.Dataset.from_tensor_slices((val_df["filepath"], val_df["label"]))
val_ds_eval = val_ds_eval.map(decode_img).batch(BATCH_SIZE)

# get predictions
preds = ft_model.predict(val_ds_eval)
y_pred = preds.argmax(axis=1)
y_true = val_df["label"].values

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix - Modified Fine-Tuned ResNet50")
plt.show()
````

````python
# =====================================
# STEP 1: Compute class weights
# =====================================

from sklearn.utils.class_weight import compute_class_weight
import numpy as np

# we have labels in train_df["label"]
class_weights_array = compute_class_weight(
    class_weight="balanced",                     # tell sklearn to auto-balance
    classes=np.unique(train_df["label"]),       # unique class labels: [0,1,2,3,4]
    y=train_df["label"].values                  # all training labels
)

# convert from array to dict form {class_index: weight}
class_weights = {int(c): float(w) for c, w in zip(np.unique(train_df["label"]), class_weights_array)}

print("Class weights being used:")
print(class_weights)
````

````python
# =====================================
# STEP 2: Data augmentation layer
# =====================================

from tensorflow.keras import layers

# this will only be active during training
data_aug = tf.keras.Sequential(
    [
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.1),   # small rotation
        layers.RandomZoom(0.1),
        layers.RandomContrast(0.1),
    ],
    name="data_augmentation"
)

print("Data augmentation layer created ")
````

````python
# =====================================
# STEP 3: Build modified ResNet50 model
# =====================================

from tensorflow.keras.applications import ResNet50
from tensorflow.keras import models

# load ResNet50 base (backbone) without top classification layer
ft_base = ResNet50(
    weights="imagenet",              # start from ImageNet pretrained weights
    include_top=False,               # we will add our own classifier head
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

# at first, we freeze the whole backbone
ft_base.trainable = False

# now define full model
inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3), name="input_image")

# apply data augmentation ONLY during training
x = data_aug(inputs)                 # no need for training=True; Keras handles it

# pass through ResNet50 backbone
x = ft_base(x, training=False)       # keep batchnorm etc. in inference mode

# convert feature maps to 1D vector
x = layers.GlobalAveragePooling2D()(x)

# add a fully connected layer for learning DR-specific features
x = layers.Dense(512, activation="relu")(x)

# dropout for regularization (reduce overfitting)
x = layers.Dropout(0.5)(x)

# final classification layer: 5 classes (04) with softmax
outputs = layers.Dense(NUM_CLASSES, activation="softmax", name="predictions")(x)

# build the full model
ft_model = models.Model(inputs, outputs, name="ResNet50_Modified")

ft_model.summary()  # optional: prints architecture
````

````python
# =====================================
# STEP 4: Stage 1 training (frozen backbone)
# =====================================

# compile with a slightly higher LR for head training
ft_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# use the same train_ds and val_ds you already created earlier
# THEY SHOULD NOT APPLY AUGMENTATION in the dataset function; augmentation is in the model now.

history_stage1 = ft_model.fit(
    train_ds,                     # training dataset (batched, shuffled)
    validation_data=val_ds,       # validation dataset
    epochs=5,                     # you can increase if training is stable
    class_weight=class_weights    # handle class imbalance
)

print("Stage 1 training completed ")
````

````python
# =====================================
# STEP 5: Fine-tune top layers of ResNet50 (Stage 2)
# =====================================

# choose how many layers to keep frozen
fine_tune_at = 100  # you can adjust: higher = more layers frozen

# unfreeze from 'fine_tune_at' to the end
for layer in ft_base.layers[:fine_tune_at]:
    layer.trainable = False
for layer in ft_base.layers[fine_tune_at:]:
    layer.trainable = True

# IMPORTANT: recompile with a LOWER learning rate for fine-tuning
ft_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# continue training
history_stage2 = ft_model.fit(
    train_ds,                     # same dataset
    validation_data=val_ds,
    epochs=10,                    # you can tweak this
    class_weight=class_weights
)

print("Stage 2 fine-tuning completed ")
````
